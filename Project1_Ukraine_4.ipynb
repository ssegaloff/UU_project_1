## Phenomena of Interest

The phenomenon being modeled is the sequential pattern of infrastructure damage events in Ukraine. The primary analytical question is whether the type of infrastructure or location of a given damage event can help predict the type of infrastructure or location of the next reported event. By analyzing the temporal sequence of damages, we aim to uncover underlying patterns that may not be apparent from aggregate statistics alone.

(Consult the data description section of this file to learn more about these phenomena and the context of this dataset.)

## Describe Your Non-Parametric Model

A first-order Markov Chain was selected as the modeling framework for its direct and interpretable approach to sequential data. The core concepts of the model are:

- **States:** Each unique `type_of_infrastructure` or unique `oblast` is treated as a distinct "state" in the model.

- **Transitions:** The model calculates the probability of moving from one state to another (e.g. a "transition"). This is based on the frequency with which one type of damage event is immediately followed by another in the chronologically sorted dataset.

- **Core Assumption:** The model operates on the "memoryless" property, meaning it assumes that the probability of the next event (the next state) depends only on the current event (the current state). It does not consider the history of events that occurred before the current one. 

This modeling approach allows us to transform the raw sequence of damage events into a matrix of transition probabilities, which forms the basis for our findings.

We did not encounter significant challenges beyond initially accessing the dataset, the need to impute dates from source_date when date_of_event was null (which we did by relying on _internal_filter_date for our primary time marker), and the missing rayon data that caused us to use the more broad oblast categories for our geographic analysis.

## Creating Functions for the Model and Simulations

```python
# Build transition matrix

def build_transition_matrix(df, sequence_col, 
                      time_col='_internal_filter_date', min_count = 10):
    """
    Build transition probability matrix from damage sequence.

    Returns: transition matrix, transition_counts, valid_states
    """
    df = df.copy()
    df[time_col] = pd.to_datetime(df[time_col], errors='coerce')
    df = df.sort_values(by=time_col).reset_index(drop=True)
    df = df[df[sequence_col].notna()].reset_index(drop=True)

    states = df[sequence_col].values

    state_counts = pd.Series(states).value_counts()
    valid_states = state_counts[state_counts >= min_count].index.tolist()

    transition_counts = pd.DataFrame(0, index = valid_states, columns = valid_states, dtype = float)

    for i in range(len(states) - 1):
        current = states[i]
        next_state = states[i + 1]
        if current in valid_states and next_state in valid_states:
            transition_counts.loc[current, next_state] += 1

    transition_matrix = transition_counts.div(transition_counts.sum(axis=1), axis=0).fillna(0)

    return transition_matrix, transition_counts, valid_states
```

```python
# Analyze transition probabilities

def get_transition_probability(transition_matrix, transition_counts, from_state, to_state):
    """Get probability of transitioning from one infrastructure type to another."""

    if from_state not in transition_matrix.index or to_state not in transition_matrix.columns:
        return None
    
    prob = transition_matrix.loc[from_state, to_state]
    count = int(transition_counts.loc[from_state, to_state])
    total = int(transition_counts.loc[from_state].sum())

    return {
        'from': from_state,
        'to': to_state,
        'probability': prob,
        'percentage': f"{prob*100:.2f}%",
        'observed_transitions': count,
        'total_from_state': total
    }

def get_most_likely_next(transition_matrix, transition_counts, current_state, top_n=5):
    """Get the most likely next damaged infrastructure types given the current damaged infrastructure type."""

    if current_state not in transition_matrix.index:
        return None
    
    transitions = transition_matrix.loc[current_state].sort_values(ascending=False).head(top_n)
    counts = transition_counts.loc[current_state]

    result = pd.DataFrame({
        'Next Infrastructure Type': transitions.index,
        'Probability': transitions.values,
        'Percentage': [f"{p*100:.2f}%" for p in transitions.values],
        'Observed Count': [int(counts[state]) for state in transitions.index]
    })

    return result
```

```python
# Calculate steady state

def calculate_steady_state(transition_matrix, max_iter=1000, tol=1e-6):
    """Calculate long-run equilibrium probabilities."""

    n = len(transition_matrix)
    state_vector = np.ones(n) / n

    for i in range(max_iter):
        new_vector = state_vector @ transition_matrix.values
        if np.abs(new_vector - state_vector).max() < tol:
            break
        state_vector = new_vector

    steady_state = pd.Series(state_vector, index=transition_matrix.index)
    return steady_state.sort_values(ascending=False)
```

```python
# Visualizations

def plot_transition_matrix(transition_matrix, transition_counts, top_n = 12):
    """Visualize transition matrix as heatmap."""

    top_states = transition_counts.sum(axis=1).nlargest(top_n).index
    matrix_subset = transition_matrix.loc[top_states, top_states]

    plt.figure(figsize=(16,14))
    sns.heatmap(matrix_subset, annot=True, fmt='.3f', cmap = 'YlOrRd',
                cbar_kws = {'label': 'Transition Probability'},
                linewidths=0.5)
    plt.title(f'Infrastructure Damage Transition Matrix (Top {top_n} Types)',
              fontsize=14, fontweight='bold')
    plt.xlabel('Next Infrastructure Type', fontsize=12)
    plt.ylabel('Current Infrastructure Type', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()


def plot_steady_state_comparison(steady_state, df, sequence_col):
    "Compare actual distribution to steady state."

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    actual = df[sequence_col].value_counts(normalize=True).head(10)
    ax1.barh(range(len(actual)), actual.values, color = 'steelblue')
    ax1.set_yticks(range(len(actual)))
    ax1.set_yticklabels(label[:40] for label in actual.index)
    ax1.set_xlabel('Proportion', fontsize=11)
    ax1.set_title('Actual Distribution\n(Observed)', fontsize=12, fontweight='bold')
    ax1.invert_yaxis()

    ax2.barh(range(len(steady_state.head(10))), steady_state.head(10).values, color = 'coral')
    ax2.set_yticks(range(len(steady_state.head(10))))
    ax2.set_yticklabels(label[:40] for label in steady_state.head(10).index)
    ax2.set_xlabel('Probability', fontsize=11)
    ax2.set_title('Steady State\n(Long-run Equilibrium)', fontsize=12, fontweight='bold')
    ax2.invert_yaxis()

    plt.tight_layout()
    plt.show()
```

```python
# Visualizations for regions (oblast)

def plot_transition_matrix_region(transition_matrix, transition_counts, top_n = 12):
    """Visualize transition matrix as heatmap."""

    top_states = transition_counts.sum(axis=1).nlargest(top_n).index
    matrix_subset = transition_matrix.loc[top_states, top_states]

    plt.figure(figsize=(16,14))
    sns.heatmap(matrix_subset, annot=True, fmt='.3f', cmap = 'crest',
                cbar_kws = {'label': 'Transition Probability'},
                linewidths=0.5)
    plt.title(f'Reginal(oblast) Transition Matrix (Top {top_n} Types)',
              fontsize=14, fontweight='bold')
    plt.xlabel('Next Region (oblast)', fontsize=12)
    plt.ylabel('Current Region (Oblast)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
def simulate_sequence(transition_matrix, initial_state, n):
    """Simulate sequence of events."""

    sequence = [initial_state]
    current_state = initial_state
    
    for i in range(n):
        next_state = np.random.choice(transition_matrix.columns, p=transition_matrix.loc[current_state])
        sequence.append(next_state)
        current_state = next_state

    return sequence
```

```python
# visualize difference between actual and simulated distributions

def plot_sim_v_actual_dist(sim_dist, df, sequence_col):
    "Compare actual distribution to simulated distributions."

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    actual = df[sequence_col].value_counts(normalize=True)
    ax1.barh(range(len(actual)), actual.values, color = 'steelblue')
    ax1.set_yticks(range(len(actual)))
    ax1.set_yticklabels(label[:40] for label in actual.index)
    ax1.set_xlabel('Proportion', fontsize=11)
    ax1.set_title('Actual Distribution', fontsize=12, fontweight='bold')
    ax1.invert_yaxis()

    ax2.barh(range(len(sim_dist)), sim_dist.values, color = 'coral')
    ax2.set_yticks(range(len(sim_dist)))
    ax2.set_yticklabels(label[:40] for label in sim_dist.index)
    ax2.set_xlabel('Probability', fontsize=11)
    ax2.set_title('Simulated Distribution', fontsize=12, fontweight='bold')
    ax2.invert_yaxis()

    plt.tight_layout()
    plt.show()
```

## Conclusion

**Limitations**

- **Model assumptions:** The modelâ€™s core assumption, that the next event depends only on the current one, oversimplifies the complex conflict dynamics. Real-world attacks are often clustered and coordinated, which is a nuance the memoryless model cannot capture.

- **Missing Data:** Significant gaps in the data, particularly missing event dates (26%) and rayon-level details (36%), reduce the temporal and spatial fidelity of the analysis. These omissions may obscure significant patterns at a more granular level.

- **Omission of Strategic Context:** The model is purely probabilistic and does not account for external strategic factors. It cannot incorporate military objectives, seasonal strategies, or tactical shifts on the ground, which are true drivers of these patterns. 

- By treating the data as a single continuous sequence, the model does not capture specific escalation phases, lulls in fighting, or seasonal variations that are visible in the raw monthly data.

**Future Work**

Building on this analysis, future work should aim to address the identified limitations to develop a more robust and operationally relevant understanding of damage patterns.

- Incorporate more complex models, such as higher-order Markov Chains or other time-series models, that can account for time-dependencies and move beyond the memoryless assumption.

- Develop models that specifically account for the geographic clustering of events. This is a critical feature of the data and is essential for building predictive models that could support operational early-warning systems for humanitarian and protection aids and networks.

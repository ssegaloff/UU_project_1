### Data Provenance: Who collected the data and why?

ACLED (Armed Conflict Location and Event Dataset) is the organization that has collected the data on the Ukraine conflict that contains the information in this dataset. They have a broader dataset on the whole of the conflict from which they pulled this data. According to the ACLED webpage for the Ukraine Conflict Monitor, the data on attacks on Ukrainian infrastructure file (ukraine-damages.csv) contains events that occured within the Ukraine that that feature at least one of the following civilian infrastructure tags: energy, health, education, and residential infrastructure.
- https://acleddata.com/methodology/ukraine-civilian-infrastructure-automated-tags
